```python
#print sum of all even and odd number in given range

n = int(input("Enter the n: "))

even = 0
odd = 0

for i in range(1,n+1):
    if i%2 == 0:
        even = even+i
    else:
        odd = odd+i

print("Sum of even is: ",even)
print("Sum of odd is: ", odd)
        
```

```python
#Print all factor of given number

n = int(input("Enter the n: "))

count = 0
for i in range(1,n-1):
    if n%i == 0:
        count = count +1
        print(i)
print(count)        
```

```python
#check given number is perfect or not

n = int(input("Enter the n: "))

sum = 0

for i in range(1,n-1):
    if n%i == 0:
        sum = sum+i
print(sum)

if sum == n:
    print("Given number is perfect number")

else:
    print("Given number is not perfect number")
```

```python
#check given number is prime or not

n = int(input("Enter the n: "))

count = 0
for i in range(1,n+1):
    if n%i == 0:
        count = count+1
        print(i)
print(count)

if count == 2:
    print("Given number is prime")
else:
    print("Given number is not prime")
```

```python
#count number of vowels in string

a = input("Enter the string: ").lower()

count = 0

for i in range(len(a)):
    if a[i] == 'a' or a[i] == 'e' or a[i] == 'i' or a[i] == 'o' or a[i] == 'u':
        count = count +1

if count == 0:
    print("No vowel is found")
else:
    print(count)

        
        
```
