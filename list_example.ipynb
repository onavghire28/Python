```python
#Write a program to find sum of element in the given list
a = list(map(int,input("enter the element: ").split()))
sum = 0

for i in range(len(a)):
    sum = sum + a[i]
print("sum is: ",sum)    
```

```python
#find largest and smallest  element in the list

a = list(map(int, input("Enter the element: ").split()))

max = a[0]
min = a[0]
for i in range(len(a)):
    if a[i] > max:
        max = a[i]
        
for i in range(len(a)):
    if a[i] < min:
        min = a[i]
        
    
print(max)
print(min)
        
```

```python
#count number of odd and even in given list

a = list(map(int, input("Enter the elements: ").split()))

even = 0
odd = 0

for i in range(len(a)):

    if a[i] % 2 == 0:
        even = even+1
    else:
        odd = odd +1
        
print(even)
print(odd)
```

```python
#Remove duplicate element from list

a = list(map(int, input("Enter the element: ").split()))

b = []

for i in range(len(a)):
    if a[i] not in b:
        b.append(a[i])
        
print(b)        
```

```python
#find the second largest element in given list
a = list(map(int, input("Enter the element: ").split()))

b = []

for i in range(len(a)):
    if a[i] not in b:
        b.append(a[i])
        
print(b)
b.sort()

print(b[-2])

```

```python
#reverse the string

a = [1,2,3,4,5,6]

for i in range(len(a)-1,-1,-1):
    print(a[i], end = " ")
```

```python
#check list is palindrome or not

lst = list(map(int, input("Enter elements: ").split()))

if lst == lst[::-1]:
    print("Palindrome list")
else:
    print("Not a palindrome list")
      
```
