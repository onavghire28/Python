```python
#checking give number is even or odd

def is_even(num):
    if type(num) == int:
        
      if num %2 == 0:
         return "even"
      else:
         return "Odd"
```

```python
is_even(10)
```

```python
for i in range(1,11):
    x = is_even(i)
    print(x)
```

```python
#Type of argument
#default argument
def add(a,b=4):
    return a+b

add(2)
```

```python
#positional argument
def x(a,b):
    return a**b

x(2,3)
```

```python
#keyword argument
def x(a,b):
    return a**b
x(a = 2, b = 3)    
```

```python
#*args
#we can pass n number of argument with help of args

def mul(*args):

    product = 1
    for i in args:
        product = product*i
    return product

mul(1,2,3,4)
mul(1,2,3)
    
    
```

```python
#nested functions

def g(x):
    def h():
        x = 'abc'
        return x
    x = x + 1
    print('in g(x): x =', x)
    h()
    return x
    

x = 3
z = g(x)
```

```python
def g(x):
    def h(x):
        x = x+1
        print("in h(x): x = ", x)
    x = x + 1
    print('in g(x): x = ', x)
    h(x)
    return x

x = 3
z = g(x)
print('in main program scope: x = ', x)
print('in main program scope: z = ', z)
```
